```python
account_details = {

}

while True :
    user_input = int(input("""
    ***********
    Select your Options :
    1) Deposit
    2) Balance Enquiry
    3) Withdrawal  
    4) Account Creation 

    Enter Your Option :    
    """))
    if(user_input == 1) :
        account_number =  int(input("""
        ***********      
        Welcome to AAA Bank
        Please Enter Your Account Number
        """))
        confirm_account_number = int(input("""
        ***********      
        Please Re-Enter Your Account Number
        """))
        if account_number == confirm_account_number :
            print(f'Hii {account_details[str(account_number)]['name']}')
            count_of_five_hundreads = int(input("""
        ***********      
        Please enter the denomination of your 500's 
        Example if you have 4 Five Hundread Rupees Note Just Enter 4
        """))
            count_of_two_hundreads = int(input("""
        ***********      
        Please enter the denomination of your 200's 
        Example if you have 4 two Hundread Rupees Note Just Enter 4
        """))
            count_of_one_hundreads = int(input("""
        ***********      
        Please enter the denomination of your 100's 
        Example if you have 4 One Hundread Rupees Note Just Enter 4
        """))
            total_amount = count_of_five_hundreads * 500 + count_of_one_hundreads * 100 + count_of_two_hundreads * 200
            confirm_amount = int(input(f"""
            ***********      
            100 * {count_of_one_hundreads} = {count_of_one_hundreads * 100}    
            200 * {count_of_two_hundreads} = {count_of_two_hundreads * 200}  
            500 * {count_of_five_hundreads} = {count_of_five_hundreads * 500} 
            Total Amount = {total_amount}  

            If the denomation is correct Please Enter 1
            Else Enter 2
            """))

            if confirm_amount == 1:
                previous_balance = account_details.get(str(account_number))['balance']
                # if previous_balance is None :
                #     account_details[str(account_number)] = total_amount
                # else :
                #     account_details[str(account_number)] = total_amount + previous_balance
                account_details[str(account_number)]['balance'] = total_amount + (0 if previous_balance is None else previous_balance)
                print(f"{total_amount} is Deposited in your account - {account_number}")
            else :
                print("Please Collect Your  Cash")          
        else :
            print("Check Your Account Number")

    elif user_input == 2 :
        account_number =  int(input("""
        ***********      
        Welcome to AAA Bank
        Please Enter Your Account Number
        """))
        confirm_account_number = int(input("""
        ***********      
        Please Re-Enter Your Account Number
        """))
        if account_number == confirm_account_number :
            print(f"Your Account Balance : {account_details[str(account_number)]['balance']}")     
        else :
            print("Account Number & Confirm Account Number is mismatch")

    elif user_input == 3 :
        account_number =  int(input("""
        ***********      
        Welcome to AAA Bank
        Please Enter Your Account Number
        """))
        confirm_account_number = int(input("""
        ***********      
        Please Re-Enter Your Account Number
        """))
        if account_number == confirm_account_number :
            withdraw_amount = int(input("""
        PLease Enter Your Withdrawal Amount : 
        """))            
            if(account_details[str(account_number)]['balance'] >= withdraw_amount) :
                account_details[str(account_number)]['balance'] -= withdraw_amount
                print(
                    f"""Withdrawal Procedure Start's
                       Please Collect Your Cash

                       *******
                        Account Number : {account_number}
                        Withdrawal Amount : {withdraw_amount}
                        Updated Balance : {account_details[str(account_number)]['balance']}
                       ********
                    """
                )
            else :
                print("There is no money in your account Bye Bye")

    elif user_input == 4 :
        name =  input("""***********      
        Welcome to AAA Bank
        Thank you for choosing Us.
        Please Enter Your Name
        """)
        age = int(input("Please enter your age :"))
        balance = int(input("Please enter your initial deposit amount :"))
        account_number = input("Enter Your Account Number :")

        person_bank_details = {
            "name" : name,
            "age" : age,
            "balance" : balance,
            "account_number" : account_number
        }

        account_details[account_number] = person_bank_details
        print("Account Created Successfully2")

    else :
        print("Please Enter Valid Input")
```
